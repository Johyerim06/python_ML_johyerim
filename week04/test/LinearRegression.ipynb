```python
import matplotlib.pyplot as plt
```

```python
import numpy as nb
import pandas as pd
```

```python
dataset = pd.read_csv("../LinearRegressionData.csv") #../ 상위디렉터리, ./ 현재디렉터리
```

```python
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
# 랜덤한한

# 나눠짐
```

```python
X_train, len(X_train)
```

```python
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression() 
reg = LinearRegression() #LinearRegression 모델 객체 생성함함
```

```python
reg # 객체가 생김 fitted 되지 않음(not fitted) - 객체만 생성됨
```

```python
reg.fit(X, y) #(fitted)
reg.fit(X_train, y_train) #(fitted)
```

```python
reg1.fit(X, y) #(fitted)
reg1.fit(X_train, y_train) #(fitted)
```

```python
reg.coef_, reg.intercept_# 10.49161294 , 직선의 방정식 값 , 학습데이터 분리 후 얻은 결과과
```

```python
reg1.coef_, reg1.intercept_
```

```python
reg.fit(X, y)
reg.coef_, reg.intercept_ # 전체 데이터로 얻은 결과
# -0.2가 되어 기울기가 조금 내려갔을 것임
reg1.fit(X, y)
reg1.coef_, reg1.intercept_
```

```python
from sklearn.linear_model import SGDRegressor
sr = SGDRegressor(max_iter = 200, eta0=0.1, random_state=0, verbose=0) #하이퍼파라미터
#max_iter = 얼마나 움직일지 바꾸는거 
#eta0 = learning rate 학습율 (조금씩 이동, 많이 이동 결정)
#sr = SGDRegressor() # 시간 돈 다 많을때 하는 방식 - 많을때 시간 겁나 많이 걸림
sr.fit(X_train, y_train)
```

```python
# 시각화
plt.scatter(X, y, color = 'r')
plt.scatter(X_train, y_train, color = 'b')
plt.plot(X, reg1.predict(X), color='r') # y로 예측된 값을 2번째 파라미터로 넣음
plt.plot(X_train, reg.predict(X_train), color='b')
plt.plot(X_train, sr.predict(X_train), color = 'cyan')
plt.show()
```

```python
reg1.score(X, y), reg.score(X_test, y_test)
```
