```python
# X = hour, y = score
X = dataset.iloc[:, :-1].values #마지막 1 줄 빼고 가져옴(feature)
y = dataset.iloc[:, -1].values # 마지막 1줄만 가져옴(lable)
# X, y 데이터 값 개수가 다르면 오류남
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() # 리니어 모델 객체를 하나 만든것임
```

```python
reg.fit(X, y) # 학습시킴 y = ax + b 에서 가
```

```python
reg.coef_, reg.intercept_ #기울기, 절편값
```

```python
# 10.44369694*10 + -0.218484702867201 기울기, 절편값 복붙함

#y_pred = reg.coef_*5.5 + reg.intercept_ #X값이 10일때 의 값
#y_pred
```

```python
y_pred = reg.predict(X)
y_pred
#Yhat(y^ 을 구해봄봄)
```

```python
plt.scatter(X, y, color = 'b')
plt.scatter(X, y_pred, color = 'r') 
#plt.plot(X, y_pred, color = 'r')  # y_pred - 선으로 나와 있는 값
plt.show()
# 선형회귀
```
